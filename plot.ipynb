```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 1: Read the calculated results from the CSV file
df = pd.read_csv('caculated_results.csv', skipinitialspace=True)
```

```python
df['method'].unique()

# rename method
df['method'] = df['method'].replace({ \
    'ilp': 'ILP', \
    'cp': 'CP', \
    'local_search_team_and_order': 'LS', \
    'tabu_search': 'TS', \
    'greedy': 'GR', \
    'GA': 'GA', \
    'local_search_time': 'LS_2', \
    'simulated_annealing': 'SA' \
})

df['method'].unique()
```

```python
df.head(n = 10)
```

```python
methods = ['ILP', 'CP', 'GR', 'LS', 'TS', 'SA', 'GA', 'LS_2']
```

```python
method_order = ['ILP', 'CP', 'GR', 'LS', 'TS', 'SA', 'GA', 'LS_2']
# Step 2: Get unique values of N
N_values = df['N'].unique()

N_values = [10, 50, 100, 500, 1000]

# Step 3: For each N, plot the avg and std for all methods
for N in N_values:
    N_data = df[df['N'] == N]

    # Create a new figure for each N
    plt.figure(figsize=(10, 6))

    # Plot avg_max_task_count with std as error bars, ensuring the method order
    for method in method_order:
        if method in N_data['method'].values:  # Check if the method is in the data for this N
            method_data = N_data[N_data['method'] == method]
            
            avg_max_task_count = method_data['avg_max_task_count'].values[0]
            std_max_task_count = method_data['std_max_task_count'].values[0]
            
            plt.errorbar(method, avg_max_task_count, yerr=std_max_task_count, label=f'Method: {method}', 
                         fmt='o', capsize=5, markersize=8)

    # Customize the plot for avg_max_task_count
    plt.title(f'Average Max Task Count with Std for N = {N}')
    plt.xlabel('Method')
    plt.ylabel('Average Max Task Count')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot avg_min_completion_time with std as error bars, ensuring the method order
    plt.figure(figsize=(10, 6))
    for method in method_order:
        if method in N_data['method'].values:  # Check if the method is in the data for this N
            method_data = N_data[N_data['method'] == method]
            
            avg_min_completion_time = method_data['avg_min_completion_time'].values[0]
            std_min_completion_time = method_data['std_min_completion_time'].values[0]
            
            plt.errorbar(method, avg_min_completion_time, yerr=std_min_completion_time, label=f'Method: {method}', 
                         fmt='o', capsize=5, markersize=8)

    # Customize the plot for avg_min_completion_time
    plt.title(f'Average Min Completion Time with Std for N = {N}')
    plt.xlabel('Method')
    plt.ylabel('Average Min Completion Time')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot avg_min_cost with std as error bars, ensuring the method order
    plt.figure(figsize=(10, 6))
    for method in method_order:
        if method in N_data['method'].values:  # Check if the method is in the data for this N
            method_data = N_data[N_data['method'] == method]
            
            avg_min_cost = method_data['avg_min_cost'].values[0]
            std_min_cost = method_data['std_min_cost'].values[0]
            
            plt.errorbar(method, avg_min_cost, yerr=std_min_cost, label=f'Method: {method}', 
                         fmt='o', capsize=5, markersize=8)

    # Customize the plot for avg_min_cost
    plt.title(f'Average Min Cost with Std for N = {N}')
    plt.xlabel('Method')
    plt.ylabel('Average Min Cost')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot std_max_task_count with error bars (optional)
    plt.figure(figsize=(10, 6))
    for method in method_order:
        if method in N_data['method'].values:  # Check if the method is in the data for this N
            method_data = N_data[N_data['method'] == method]
            
            std_max_task_count = method_data['std_max_task_count'].values[0]
            
            plt.plot(method, std_max_task_count, 'o', label=f'Method: {method}', markersize=8)

    # Customize the plot for std_max_task_count
    plt.title(f'Standard Deviation of Max Task Count for N = {N}')
    plt.xlabel('Method')
    plt.ylabel('Standard Deviation of Max Task Count')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
